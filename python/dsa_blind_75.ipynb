## Array Problems

### Two Sum

```python
def two_sum(nums, target):
    seen = {}
    for i, v in enumerate(nums):
        need = target - v
        if need in seen:
            return [seen[need], i]
        seen[v] = i

print('two_sum([2,7,11,15], 9):', two_sum([2,7,11,15], 9))
```

```python
# Best Time to Buy and Sell Stock
def max_profit(prices):
    min_p = float('inf')
    profit = 0
    for p in prices:
        min_p = min(min_p, p)
        profit = max(profit, p - min_p)
    return profit

print('max_profit([7,1,5,3,6,4]):', max_profit([7,1,5,3,6,4]))
```

### Contains Duplicate

```python
def product_except_self(nums):
    n = len(nums)
    out = [1]*n
    prefix = 1
    for i in range(n):
        out[i] = prefix
        prefix *= nums[i]
    suffix = 1
    for i in range(n-1, -1, -1):
        out[i] *= suffix
        suffix *= nums[i]
    return out

print('product_except_self([1,2,3,4]):', product_except_self([1,2,3,4]))
```

### Product of Array Except Self

```python
def max_subarray(nums):
    best = cur = nums[0]
    for x in nums[1:]:
        cur = max(x, cur + x)
        best = max(best, cur)
    return best

print('max_subarray([-2,1,-3,4,-1,2,1,-5,4]):', max_subarray([-2,1,-3,4,-1,2,1,-5,4]))
```

### Maximum Subarray (Kadane's Algorithm)

```python
def max_product_subarray(nums):
    cur_max = cur_min = ans = nums[0]
    for x in nums[1:]:
        if x < 0:
            cur_max, cur_min = cur_min, cur_max
        cur_max = max(x, cur_max * x)
        cur_min = min(x, cur_min * x)
        ans = max(ans, cur_max)
    return ans

print('max_product_subarray([2,3,-2,4]):', max_product_subarray([2,3,-2,4]))
```

### Maximum Product Subarray

```python
def find_min_rotated(nums):
    lo, hi = 0, len(nums)-1
    while lo < hi:
        mid = (lo + hi)//2
        if nums[mid] > nums[hi]:
            lo = mid + 1
        else:
            hi = mid
    return nums[lo]

print('find_min_rotated([4,5,6,7,0,1,2]):', find_min_rotated([4,5,6,7,0,1,2]))
```

### Find Minimum in Rotated Sorted Array

```python
def search_rotated(nums, target):
    lo, hi = 0, len(nums)-1
    while lo <= hi:
        mid = (lo + hi)//2
        if nums[mid] == target:
            return mid
        if nums[lo] <= nums[mid]:
            if nums[lo] <= target < nums[mid]:
                hi = mid - 1
            else:
                lo = mid + 1
        else:
            if nums[mid] < target <= nums[hi]:
                lo = mid + 1
            else:
                hi = mid - 1
    return -1

print('search_rotated([4,5,6,7,0,1,2], 0):', search_rotated([4,5,6,7,0,1,2], 0))
```

### Search in Rotated Sorted Array

```python
def three_sum(nums):
    nums.sort()
    res = []
    n = len(nums)
    for i in range(n-2):
        if i and nums[i] == nums[i-1]:
            continue
        l, r = i+1, n-1
        while l < r:
            s = nums[i] + nums[l] + nums[r]
            if s < 0: l += 1
            elif s > 0: r -= 1
            else:
                res.append([nums[i], nums[l], nums[r]])
                while l < r and nums[l] == nums[l+1]: l += 1
                while l < r and nums[r] == nums[r-1]: r -= 1
                l += 1; r -= 1
    return res

print('three_sum([-1,0,1,2,-1,-4]):', three_sum([-1,0,1,2,-1,-4]))
```

### 3 Sum

```python
def max_area(height):
    l, r, ans = 0, len(height)-1, 0
    while l < r:
        ans = max(ans, min(height[l], height[r])*(r-l))
        if height[l] < height[r]: l += 1
        else: r -= 1
    return ans

print('max_area([1,8,6,2,5,4,8,3,7]):', max_area([1,8,6,2,5,4,8,3,7]))
```

## Binary Problems

### Sum of Two Integers

```python
def get_sum(a, b):
    MASK = 0xFFFFFFFF
    while b != 0:
        a, b = (a ^ b) & MASK, ((a & b) << 1) & MASK
    return a if a <= 0x7FFFFFFF else ~(a ^ MASK)

print('get_sum(1, 2):', get_sum(1, 2))
```

### Number of 1 Bits

```python
def hamming_weight(n):
    count = 0
    while n:
        n &= n-1
        count += 1
    return count

print('hamming_weight(11):', hamming_weight(11))
```

### Counting Bits

```python
def count_bits(n):
    res = [0]*(n+1)
    for i in range(1, n+1):
        res[i] = res[i & (i-1)] + 1
    return res

print('count_bits(5):', count_bits(5))
```

### Missing Number

```python
def missing_number(nums):
    n = len(nums)
    res = n
    for i, num in enumerate(nums):
        res ^= i ^ num
    return res

print('missing_number([3,0,1]):', missing_number([3,0,1]))
```

### Reverse Bits

```python
def reverse_bits(n):
    res = 0
    for _ in range(32):
        res = (res << 1) | (n & 1)
        n >>= 1
    return res

print('reverse_bits(43261596):', reverse_bits(43261596))
```

## Dynamic Programming

### Climbing Stairs

```python
def climb_stairs(n):
    a, b = 1, 1
    for _ in range(n-1):
        a, b = b, a+b
    return b

print('climb_stairs(5):', climb_stairs(5))
```

### Coin Change

```python
def coin_change(coins, amount):
    dp = [float('inf')] * (amount+1)
    dp[0] = 0
    for c in coins:
        for x in range(c, amount+1):
            dp[x] = min(dp[x], dp[x-c]+1)
    return dp[amount] if dp[amount] != float('inf') else -1

print('coin_change([1,2,5], 11):', coin_change([1,2,5], 11))
```

### Longest Increasing Subsequence

```python
import bisect
def length_of_lis(nums):
    dp = []
    for x in nums:
        i = bisect.bisect_left(dp, x)
        if i == len(dp): dp.append(x)
        else: dp[i] = x
    return len(dp)

print('length_of_lis([10,9,2,5,3,7,101,18]):', length_of_lis([10,9,2,5,3,7,101,18]))
```

### Longest Common Subsequence

```python
def lcs(a, b):
    m, n = len(a), len(b)
    dp = [[0]*(n+1) for _ in range(m+1)]
    for i in range(m-1, -1, -1):
        for j in range(n-1, -1, -1):
            if a[i] == b[j]: dp[i][j] = 1 + dp[i+1][j+1]
            else: dp[i][j] = max(dp[i+1][j], dp[i][j+1])
    return dp[0][0]

print('lcs("abcde", "ace"):', lcs('abcde', 'ace'))
```

### Word Break

```python
def word_break(s, wordDict):
    dp = [False]*(len(s)+1)
    dp[0] = True
    for i in range(1, len(s)+1):
        for j in range(i):
            if dp[j] and s[j:i] in wordDict:
                dp[i] = True
                break
    return dp[-1]

print('word_break("leetcode", {"leet","code"}):', word_break('leetcode', {'leet','code'}))
```

### Combination Sum

```python
def combination_sum(candidates, target):
    res = []
    candidates.sort()
    def backtrack(start, path, tot):
        if tot == target:
            res.append(path[:])
            return
        for i in range(start, len(candidates)):
            c = candidates[i]
            if tot + c > target: break
            path.append(c)
            backtrack(i, path, tot+c)
            path.pop()
    backtrack(0, [], 0)
    return res

print('combination_sum([2,3,6,7], 7):', combination_sum([2,3,6,7], 7))
```

### House Robber

```python
def rob(nums):
    prev, cur = 0, 0
    for n in nums:
        prev, cur = cur, max(cur, prev + n)
    return cur

print('rob([1,2,3,1]):', rob([1,2,3,1]))
```

### House Robber II

```python
def rob_circle(nums):
    if len(nums) == 1: return nums[0]
    def rob_linear(arr):
        prev, cur = 0, 0
        for x in arr:
            prev, cur = cur, max(cur, prev + x)
        return cur
    return max(rob_linear(nums[:-1]), rob_linear(nums[1:]))

print('rob_circle([2,3,2]):', rob_circle([2,3,2]))
```

### Decode Ways

```python
def num_decodings(s):
    if not s or s[0]=='0': return 0
    n = len(s)
    dp = [0]*(n+1)
    dp[0] = 1
    for i in range(1, n+1):
        if s[i-1] != '0': dp[i] += dp[i-1]
        if i>1 and 10 <= int(s[i-2:i]) <= 26: dp[i] += dp[i-2]
    return dp[n]

print('num_decodings("226"):', num_decodings('226'))
```

## Graph Problems

### Clone Graph

```python
class Node:
    def __init__(self, val=0, neighbors=None):
        self.val = val
        self.neighbors = neighbors or []

def clone_graph(node):
    if not node: return None
    from collections import deque
    old_to_new = {node: Node(node.val)}
    q = deque([node])
    while q:
        n = q.popleft()
        for nei in n.neighbors:
            if nei not in old_to_new:
                old_to_new[nei] = Node(nei.val)
                q.append(nei)
            old_to_new[n].neighbors.append(old_to_new[nei])
    return old_to_new[node]

print('clone_graph: function defined')
```

### Course Schedule

```python
def can_finish(numCourses, prerequisites):
    from collections import defaultdict, deque
    indeg = [0]*numCourses
    g = defaultdict(list)
    for a,b in prerequisites:
        g[b].append(a)
        indeg[a] += 1
    q = deque([i for i in range(numCourses) if indeg[i]==0])
    cnt = 0
    while q:
        u = q.popleft(); cnt += 1
        for v in g[u]:
            indeg[v]-=1
            if indeg[v]==0: q.append(v)
    return cnt == numCourses

print('can_finish(2, [[1,0]]):', can_finish(2, [[1,0]]))
```

### Number of Islands

```python
def num_islands(grid):
    if not grid: return 0
    m, n = len(grid), len(grid[0])
    def dfs(i,j):
        if i<0 or j<0 or i>=m or j>=n or grid[i][j]=='0': return
        grid[i][j] = '0'
        for di,dj in [(1,0),(-1,0),(0,1),(0,-1)]:
            dfs(i+di,j+dj)
    cnt = 0
    for i in range(m):
        for j in range(n):
            if grid[i][j] == '1':
                cnt += 1
                dfs(i,j)
    return cnt

print('num_islands: function defined')
```

### Longest Consecutive Sequence

```python
def longest_consecutive(nums):
    s = set(nums)
    best = 0
    for x in s:
        if x-1 not in s:
            cur = 1
            while x+cur in s: cur += 1
            best = max(best, cur)
    return best

print('longest_consecutive([100,4,200,1,3,2]):', longest_consecutive([100,4,200,1,3,2]))
```

### Number of Connected Components

```python
def count_components(n, edges):
    parent = list(range(n))
    def find(x):
        while parent[x]!=x:
            parent[x] = parent[parent[x]]
            x = parent[x]
        return x
    def union(a,b):
        pa, pb = find(a), find(b)
        if pa!=pb: parent[pa]=pb
    for a,b in edges: union(a,b)
    return len({find(i) for i in range(n)})

print('count_components(5, [[0,1],[1,2],[3,4]]):', count_components(5, [[0,1],[1,2],[3,4]]))
```

### Pacific Atlantic Water Flow

```python
def pacific_atlantic(heights):
    if not heights: return []
    m, n = len(heights), len(heights[0])
    pac, atl = set(), set()
    
    def dfs(i, j, visited, prev):
        if (i, j) in visited or i < 0 or j < 0 or i >= m or j >= n or heights[i][j] < prev:
            return
        visited.add((i, j))
        for di, dj in [(0, 1), (1, 0), (0, -1), (-1, 0)]:
            dfs(i + di, j + dj, visited, heights[i][j])
    
    for i in range(m):
        dfs(i, 0, pac, 0)
        dfs(i, n - 1, atl, 0)
    for j in range(n):
        dfs(0, j, pac, 0)
        dfs(m - 1, j, atl, 0)
    
    return list(pac & atl)

print('pacific_atlantic: function defined')
```

### Alien Dictionary

```python
def alien_order(words):
    from collections import defaultdict, deque
    g = defaultdict(set)
    indeg = {c: 0 for w in words for c in w}
    for i in range(len(words) - 1):
        w1, w2 = words[i], words[i + 1]
        for j in range(min(len(w1), len(w2))):
            if w1[j] != w2[j]:
                if w2[j] not in g[w1[j]]:
                    g[w1[j]].add(w2[j])
                    indeg[w2[j]] += 1
                break
    q = deque([c for c in indeg if indeg[c] == 0])
    res = []
    while q:
        c = q.popleft()
        res.append(c)
        for nei in g[c]:
            indeg[nei] -= 1
            if indeg[nei] == 0: q.append(nei)
    return ''.join(res) if len(res) == len(indeg) else ''

print('alien_order: function defined')
```

## Interval Problems

### Insert Interval

```python
def insert_interval(intervals, newInterval):
    res = []
    i = 0
    while i < len(intervals) and intervals[i][1] < newInterval[0]:
        res.append(intervals[i]); i+=1
    while i < len(intervals) and intervals[i][0] <= newInterval[1]:
        newInterval = [min(newInterval[0], intervals[i][0]), max(newInterval[1], intervals[i][1])]; i+=1
    res.append(newInterval)
    while i < len(intervals): res.append(intervals[i]); i+=1
    return res

print('insert_interval: function defined')
```

### Merge Intervals

```python
def merge_intervals(intervals):
    intervals.sort()
    res = []
    for it in intervals:
        if not res or res[-1][1] < it[0]: res.append(it)
        else: res[-1][1] = max(res[-1][1], it[1])
    return res

print('merge_intervals([[1,3],[2,6],[8,10],[15,18]]):', merge_intervals([[1,3],[2,6],[8,10],[15,18]]))
```

### Non-overlapping Intervals

```python
def erase_overlap_intervals(intervals):
    if not intervals: return 0
    intervals.sort(key=lambda x: x[1])
    end = intervals[0][1]
    cnt = 0
    for s,e in intervals[1:]:
        if s < end:
            cnt += 1
        else:
            end = e
    return cnt

print('erase_overlap_intervals([[1,2],[2,3],[3,4],[1,3]]):', erase_overlap_intervals([[1,2],[2,3],[3,4],[1,3]]))
```

### Meeting Rooms

```python
def can_attend_meetings(intervals):
    intervals.sort()
    for i in range(1, len(intervals)):
        if intervals[i][0] < intervals[i-1][1]: return False
    return True

print('can_attend_meetings: function defined')
```

### Meeting Rooms II

```python
import heapq
def min_meeting_rooms(intervals):
    if not intervals: return 0
    intervals.sort()
    heap = []
    heapq.heappush(heap, intervals[0][1])
    for s,e in intervals[1:]:
        if s >= heap[0]: heapq.heappop(heap)
        heapq.heappush(heap, e)
    return len(heap)

print('min_meeting_rooms([[0,30],[5,10],[15,20]]):', min_meeting_rooms([[0,30],[5,10],[15,20]]))
```

## Linked List Problems

### Reverse Linked List

```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

def reverse_list(head):
    prev = None
    cur = head
    while cur:
        nxt = cur.next
        cur.next = prev
        prev = cur
        cur = nxt
    return prev

print('reverse_list: function defined')
```

### Detect Cycle in Linked List

```python
def has_cycle(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next; fast = fast.next.next
        if slow is fast: return True
    return False

print('has_cycle: function defined')
```

### Merge Two Sorted Lists

```python
def merge_two_lists(l1, l2):
    dummy = ListNode(0); cur = dummy
    while l1 and l2:
        if l1.val < l2.val: cur.next = l1; l1 = l1.next
        else: cur.next = l2; l2 = l2.next
        cur = cur.next
    cur.next = l1 or l2
    return dummy.next

print('merge_two_lists: function defined')
```

### Merge K Sorted Lists

```python
def merge_k_lists(lists):
    import heapq
    heap = []
    for i, node in enumerate(lists):
        if node: heapq.heappush(heap, (node.val, i, node))
    dummy = ListNode(0); cur = dummy
    while heap:
        val, i, node = heapq.heappop(heap)
        cur.next = node
        cur = cur.next
        if node.next: heapq.heappush(heap, (node.next.val, i, node.next))
    return dummy.next

print('merge_k_lists: function defined')
```

### Remove Nth Node From End

```python
def remove_nth_from_end(head, n):
    dummy = ListNode(0); dummy.next = head
    first = second = dummy
    for _ in range(n+1): first = first.next
    while first:
        first = first.next; second = second.next
    second.next = second.next.next
    return dummy.next

print('remove_nth_from_end: function defined')
```

### Reorder List

```python
def reorder_list(head):
    if not head: return
    slow = fast = head
    while fast and fast.next:
        slow = slow.next; fast = fast.next.next
    prev, cur = None, slow.next
    slow.next = None
    while cur:
        nxt = cur.next
        cur.next = prev
        prev = cur; cur = nxt
    first, second = head, prev
    while second:
        fnext, snext = first.next, second.next
        first.next = second
        second.next = fnext
        first, second = fnext, snext

print('reorder_list: function defined')
```

## Matrix Problems

### Set Matrix Zeroes

```python
def set_matrix_zeroes(matrix):
    m, n = len(matrix), len(matrix[0])
    rows = set(); cols = set()
    for i in range(m):
        for j in range(n):
            if matrix[i][j]==0: rows.add(i); cols.add(j)
    for i in range(m):
        for j in range(n):
            if i in rows or j in cols: matrix[i][j]=0

print('set_matrix_zeroes: function defined')
```

### Spiral Matrix

```python
def spiral_order(matrix):
    res = []
    while matrix:
        res += matrix.pop(0)
        matrix = list(zip(*matrix))[::-1] if matrix else []
    return res

print('spiral_order: function defined')
```

### Rotate Image

```python
def rotate_image(a):
    n = len(a)
    for i in range(n):
        for j in range(i, n):
            a[i][j], a[j][i] = a[j][i], a[i][j]
    for i in range(n):
        a[i].reverse()

print('rotate_image: function defined')
```

### Word Search

```python
def exist(board, word):
    m, n = len(board), len(board[0])
    def dfs(i,j,k):
        if k == len(word): return True
        if i<0 or j<0 or i>=m or j>=n or board[i][j] != word[k]: return False
        tmp = board[i][j]; board[i][j] = '#'
        for di,dj in [(1,0),(-1,0),(0,1),(0,-1)]:
            if dfs(i+di,j+dj,k+1):
                board[i][j] = tmp; return True
        board[i][j] = tmp
        return False
    for i in range(m):
        for j in range(n):
            if dfs(i,j,0): return True
    return False

print('exist: function defined')
```

## String Problems

### Longest Substring Without Repeating Characters

```python
def length_of_longest_substring(s):
    seen = {}
    start = 0; best = 0
    for i, ch in enumerate(s):
        if ch in seen and seen[ch] >= start:
            start = seen[ch] + 1
        seen[ch] = i
        best = max(best, i-start+1)
    return best

print('length_of_longest_substring("abcabcbb"):', length_of_longest_substring('abcabcbb'))
```

### Longest Repeating Character Replacement

```python
from collections import defaultdict
def character_replacement(s, k):
    count = defaultdict(int)
    left = 0; maxf = 0; res = 0
    for right in range(len(s)):
        count[s[right]] += 1
        maxf = max(maxf, count[s[right]])
        while (right-left+1) - maxf > k:
            count[s[left]] -= 1; left += 1
        res = max(res, right-left+1)
    return res

print('character_replacement("ABAB", 2):', character_replacement('ABAB', 2))
```

### Minimum Window Substring

```python
from collections import Counter
def min_window(s, t):
    need = Counter(t)
    missing = len(t)
    i = start = end = 0
    for j, ch in enumerate(s, 1):
        if need[ch] > 0: missing -= 1
        need[ch] -= 1
        if missing == 0:
            while i < j and need[s[i]] < 0:
                need[s[i]] += 1; i += 1
            if end == 0 or j - i < end - start:
                start, end = i, j
            need[s[i]] += 1; i += 1; missing = 1
    return s[start:end] if end else ''

print('min_window("ADOBECODEBANC", "ABC"):', min_window('ADOBECODEBANC','ABC'))
```

### Valid Anagram

### Group Anagrams

```python
from collections import defaultdict
def group_anagrams(strs):
    d = defaultdict(list)
    for s in strs:
        d[tuple(sorted(s))].append(s)
    return list(d.values())

print('group_anagrams(["eat","tea","tan","ate","nat","bat"]):', group_anagrams(['eat','tea','tan','ate','nat','bat']))
```

### Valid Parentheses

```python
def valid_parentheses(s):
    stack = []
    m = {')':'(', ']':'[', '}':'{'}
    for ch in s:
        if ch in m:
            if not stack or stack.pop() != m[ch]: return False
        else: stack.append(ch)
    return not stack

print('valid_parentheses("()"):', valid_parentheses('()'))
```

### Longest Palindromic Substring

```python
def longest_palindrome(s):
    if not s: return ''
    start = 0; maxlen = 1
    def expand(l, r):
        while l >= 0 and r < len(s) and s[l] == s[r]:
            l -= 1; r += 1
        return l + 1, r
    for i in range(len(s)):
        l1, r1 = expand(i, i)
        l2, r2 = expand(i, i + 1)
        if r1 - l1 > maxlen:
            maxlen = r1 - l1; start = l1
        if r2 - l2 > maxlen:
            maxlen = r2 - l2; start = l2
    return s[start:start + maxlen]

print('longest_palindrome("babad"):', longest_palindrome('babad'))
```

### Valid Palindrome

```python
def is_palindrome(s):
    l, r = 0, len(s) - 1
    while l < r:
        while l < r and not s[l].isalnum(): l += 1
        while l < r and not s[r].isalnum(): r -= 1
        if s[l].lower() != s[r].lower(): return False
        l += 1; r -= 1
    return True

print('is_palindrome("A man, a plan, a canal: Panama"):', is_palindrome('A man, a plan, a canal: Panama'))
```

### Palindromic Substrings

```python
def count_substrings(s):
    count = 0
    for i in range(len(s)):
        for l, r in [(i, i), (i, i + 1)]:
            while l >= 0 and r < len(s) and s[l] == s[r]:
                count += 1
                l -= 1; r += 1
    return count

print('count_substrings("abc"):', count_substrings('abc'))
```

## Tree Problems

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

### Max Depth of Binary Tree

```python
def max_depth(root):
    if not root: return 0
    return 1 + max(max_depth(root.left), max_depth(root.right))

# Test
root = TreeNode(3)
root.left = TreeNode(9)
root.right = TreeNode(20)
root.right.left = TreeNode(15)
root.right.right = TreeNode(7)
print('max_depth(root):', max_depth(root))
```

### Same Tree

```python
def is_same_tree(p, q):
    if not p and not q: return True
    if not p or not q: return False
    return p.val == q.val and is_same_tree(p.left, q.left) and is_same_tree(p.right, q.right)

# Test
p = TreeNode(1)
p.left = TreeNode(2)
p.right = TreeNode(3)

q = TreeNode(1)
q.left = TreeNode(2)
q.right = TreeNode(3)

print('is_same_tree(p, q):', is_same_tree(p, q))
```

### Invert Binary Tree

```python
def invert_tree(root):
    if not root: return None
    root.left, root.right = invert_tree(root.right), invert_tree(root.left)
    return root

# Test
root = TreeNode(2)
root.left = TreeNode(1)
root.right = TreeNode(3)
inverted = invert_tree(root)
print('inverted tree root val:', inverted.val, 'left:', inverted.left.val, 'right:', inverted.right.val)
```

### Max Path Sum

```python
def max_path_sum(root):
    res = [float('-inf')]
    def dfs(node):
        if not node: return 0
        l = max(dfs(node.left), 0)
        r = max(dfs(node.right), 0)
        res[0] = max(res[0], node.val + l + r)
        return node.val + max(l, r)
    dfs(root)
    return res[0]

# Test
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
print('max_path_sum(root):', max_path_sum(root))
```

### Serialize and Deserialize Binary Tree

```python
from collections import deque

def serialize(root):
    res = []
    def dfs(node):
        if not node:
            res.append('None')
            return
        res.append(str(node.val))
        dfs(node.left)
        dfs(node.right)
    dfs(root)
    return ','.join(res)

def deserialize(data):
    vals = iter(data.split(','))
    def dfs():
        val = next(vals)
        if val == 'None': return None
        node = TreeNode(int(val))
        node.left = dfs()
        node.right = dfs()
        return node
    return dfs()

# Test
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
serialized = serialize(root)
print('serialized:', serialized)
deserialized = deserialize(serialized)
print('deserialized root val:', deserialized.val)
```

### Binary Tree Level Order Traversal

```python
from collections import deque
def level_order(root):
    if not root: return []
    res, q = [], deque([root])
    while q:
        level = []
        for _ in range(len(q)):
            node = q.popleft()
            level.append(node.val)
            if node.left: q.append(node.left)
            if node.right: q.append(node.right)
        res.append(level)
    return res

print('level_order: function defined')
```

### Subtree of Another Tree

```python
def is_subtree(root, subRoot):
    def is_same(p, q):
        if not p and not q: return True
        if not p or not q: return False
        return p.val == q.val and is_same(p.left, q.left) and is_same(p.right, q.right)
    
    if not root: return False
    if is_same(root, subRoot): return True
    return is_subtree(root.left, subRoot) or is_subtree(root.right, subRoot)

print('is_subtree: function defined')
```

### Construct Binary Tree from Preorder and Inorder

```python
def build_tree(preorder, inorder):
    if not preorder: return None
    root_val = preorder[0]
    root = TreeNode(root_val)
    mid = inorder.index(root_val)
    root.left = build_tree(preorder[1:mid+1], inorder[:mid])
    root.right = build_tree(preorder[mid+1:], inorder[mid+1:])
    return root

print('build_tree: function defined')
```

### Validate Binary Search Tree

```python
def is_valid_bst(root):
    def dfs(node, lo, hi):
        if not node: return True
        if node.val <= lo or node.val >= hi: return False
        return dfs(node.left, lo, node.val) and dfs(node.right, node.val, hi)
    return dfs(root, float('-inf'), float('inf'))

print('is_valid_bst: function defined')
```

### Kth Smallest Element in a BST

```python
def kth_smallest(root, k):
    stack = []
    node = root
    while stack or node:
        while node:
            stack.append(node)
            node = node.left
        node = stack.pop()
        k -= 1
        if k == 0: return node.val
        node = node.right

print('kth_smallest: function defined')
```

```python
from collections import Counter
import heapq

def top_k_frequent(nums, k):
    count = Counter(nums)
    return [x[0] for x in count.most_common(k)]

# Test
nums = [1, 1, 1, 2, 2, 3]
k = 2
print('top_k_frequent([1,1,1,2,2,3], 2):', top_k_frequent(nums, k))
```

### Median Finder (Find Median from Data Stream)

```python
import heapq

class MedianFinder:
    def __init__(self):
        self.small = []  # max heap (use negative for Python's min heap)
        self.large = []  # min heap
    
    def addNum(self, num):
        if not self.small or num <= -self.small[0]:
            heapq.heappush(self.small, -num)
        else:
            heapq.heappush(self.large, num)
        
        # balance: small should have at most 1 more element
        if len(self.small) > len(self.large) + 1:
            heapq.heappush(self.large, -heapq.heappop(self.small))
        if len(self.large) > len(self.small):
            heapq.heappush(self.small, -heapq.heappop(self.large))
    
    def findMedian(self):
        if len(self.small) > len(self.large):
            return float(-self.small[0])
        return (-self.small[0] + self.large[0]) / 2

# Test
mf = MedianFinder()
mf.addNum(1)
print('median after adding 1:', mf.findMedian())
mf.addNum(2)
print('median after adding 2:', mf.findMedian())
mf.addNum(3)
print('median after adding 3:', mf.findMedian())
```

## Heap Problems

#### Merge K Sorted Lists (Hard)

```python
from typing import List, Optional
import heapq

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:

        heap = []

        for i, node in enumerate(lists):
            if node:
                heapq.heappush(heap, (node.val, i, node))

        dummy = ListNode(0)
        curr = dummy

        while heap:
            val, i, node = heapq.heappop(heap)

            curr.next = node
            curr = curr.next

            if node.next:
                heapq.heappush(heap, (node.next.val, i, node.next))

        return dummy.next
```

#### 347. Top K Frequent Elements (Medium)

```python
from typing import List
from collections import Counter

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:

        count = Counter(nums)

        freq = [[] for _ in range(len(nums) + 1)]

        for num, c in count.items():
            freq[c].append(num)

        res = []

        for i in range(len(freq) - 1, 0, -1):
            for num in freq[i]:
                res.append(num)

                if len(res) == k:
                    return res
```

####  Find Median from Data Stream (Hard)

```python
import heapq

class MedianFinder:

    def __init__(self):
        self.small = []   # max heap
        self.large = []   # min heap

    def addNum(self, num: int) -> None:

        heapq.heappush(self.small, -num)

        if self.small and self.large and (-self.small[0] > self.large[0]):
            val = -heapq.heappop(self.small)
            heapq.heappush(self.large, val)

        if len(self.small) > len(self.large) + 1:
            val = -heapq.heappop(self.small)
            heapq.heappush(self.large, val)

        if len(self.large) > len(self.small) + 1:
            val = heapq.heappop(self.large)
            heapq.heappush(self.small, -val)

    def findMedian(self) -> float:

        if len(self.small) > len(self.large):
            return -self.small[0]

        if len(self.large) > len(self.small):
            return self.large[0]

        return (-self.small[0] + self.large[0]) / 2
```
